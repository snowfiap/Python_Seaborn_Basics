# Python - Seaborn

```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

## Line Plot

```python
fmri=sns.load_dataset("fmri")
fmri.head(10)
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri)
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",hue="event",data=fmri,style="event",markers=True)
plt.show()
```

## Bar Chart

```python
pokemon=pd.read_csv("pokemon.csv")
pokemon.head()
```

```python
sns.barplot(x="HP",y="Speed",data=pokemon,color="Red")
plt.show()
```

```python
sns.barplot(x="Generation",y="Legendary",data=pokemon)
plt.show()
```

```python
sns.barplot(x="Legendary",y="Speed",hue="Generation",data=pokemon)
plt.show()
```

## Scatter Plot

```python
iris=pd.read_csv("/content/iris.csv")
iris.head()
```

```python
plt.scatter(x="sepal.length",y="petal.length",data=iris)
plt.show()
```

```python
sns.scatterplot(x="sepal.length",y="petal.length",data=iris,hue="variety",style="variety")
plt.show()
```

```python
sns.distplot(pokemon["Generation"])
plt.show()
```

```python
#To get the frequency line without histogram
sns.distplot(pokemon["Generation"],hist=False)
```

```python
#Changing the color of the graph without frequency graph
sns.distplot(pokemon["Generation"],color="Red",bins=15,kde=False)
plt.show()
```

```python
#Changing the color of the graph without frequency graph
sns.distplot(pokemon["Generation"],color="Green",bins=20,vertical=True)
plt.show()
```

## Joint Plot

```python
sns.jointplot(x="sepal.length",y="petal.length",data=iris)
plt.show()
```

```python
#changing the color of the jointplot
sns.jointplot(x="sepal.length",y="petal.length",data=iris,color="pink")
plt.show()
```

```python
#Drawing a linear regression line
sns.jointplot(x="sepal.length",y="petal.length",data=iris,color="olive",kind="reg")
plt.show()
```

## Box Plot

```python
churn=pd.read_csv("/content/churn.csv")
churn.head()
```

```python
sns.boxplot(y="Intl Calls",x="Churn?",data=churn)
plt.show()
```

```python
sns.boxplot(y="Intl Calls",x="Churn?",data=churn,palette="Set1")
plt.show()
```
